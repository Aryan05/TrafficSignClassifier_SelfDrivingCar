````python
# Perform Data Augmentation
IMAGE_SIZE = 32

def centrally_scale_images(X_imgs):
    '''
    Centrally scale the images using TF
    :param X_imgs: List of images represented as a ndarray
    :return List of scaled images, each represented an a ndarray of dtype=np.uint8 [0-255]
    '''
    # List of ratios that determine the size of the bounding box
    # over which to perform the image crop and scaling operation
    scales = np.linspace(0.6, 1.1, 5)
    
    # Calculating tf.image.crop_and_resize() params
    boxes = np.zeros([len(scales), 4], dtype=np.float32)
    for index, scale in enumerate(scales):
        x1 = y1 = 0.5 - 0.5 * scale # To scale centrally
        x2 = y2 = 0.5 + 0.5 * scale
        boxes[index] = np.array([y1, x1, y2, x2], dtype=np.float32)

    box_ind = np.zeros(len(scales), dtype=np.int32)
    crop_size = np.array([IMAGE_SIZE, IMAGE_SIZE], dtype=np.int32)

    X_scale_data = []
    tf.reset_default_graph()
    X = tf.placeholder(tf.float32, shape=[1, IMAGE_SIZE, IMAGE_SIZE, 3])
    tf_img = tf.image.crop_and_resize(X, boxes, box_ind, crop_size)

    with tf.Session() as session:
        for img in X_imgs:
            batch_img = np.expand_dims(img, axis=0)
            scaled_imgs = session.run(tf_img, feed_dict={X: batch_img})
            X_scale_data.extend(scaled_imgs)

        X_scale_data = np.array(X_scale_data, dtype=np.uint8)
        return X_scale_data

def get_translate_params(index): 
    '''
    Defines the size and location of the offset window over the input image from which we
    extract glimpses
    :param index: Enum value representing the direction of image translation
    :return Tuple (coordinates (y,x) of the center of the 'offset' window
                , 'size' of the offset window
                , starting x coord for translated img
                , ending x coord for translated img
                , starting y coord for translated img
                , starting y coord for translated img
                )
    '''
    if index == 0: # Translate left 20%
        offset = np.array([0.0, 0.2], dtype=np.float32)
        size = np.array([IMAGE_SIZE, math.ceil(0.8 * IMAGE_SIZE)], dtype=np.float32)
        w_start = 0
        w_end = int(math.ceil(0.8 * IMAGE_SIZE))
        h_start = 0
        h_end = IMAGE_SIZE
    elif index == 1: # Translate right 20%
        offset = np.array([0.0, -0.2], dtype=np.float32)
        size = np.array([IMAGE_SIZE, math.ceil(0.8 * IMAGE_SIZE)], dtype=np.float32)
        w_start = int(math.floor(0.2 * IMAGE_SIZE))
        w_end = IMAGE_SIZE
        h_start = 0
        h_end = IMAGE_SIZE
    elif index == 2: # Translate top 20%
        offset = np.array([0.2, 0.0], dtype=np.float32)
        size = np.array([math.ceil(0.8 * IMAGE_SIZE), IMAGE_SIZE], dtype=np.float32)
        w_start = 0
        w_end = IMAGE_SIZE
        h_start = 0
        h_end = int(math.ceil(0.8 * IMAGE_SIZE))    
    elif index == 3: # Translate bottom 20%
        offset = np.array([-0.2, 0.0], dtype=np.float32)
        size = np.array([math.ceil(0.8 * IMAGE_SIZE), IMAGE_SIZE], dtype=np.float32)
        w_start = 0
        w_end = IMAGE_SIZE
        h_start = int(math.floor(0.2 * IMAGE_SIZE))
        h_end = IMAGE_SIZE 
        
    return offset, size, w_start, w_end, h_start, h_end

def translate_images(X_imgs):
    '''
    Performs translation of images along both the horizontal and vertical axes
    :param X_imgs: List of images, each represented as a ndarray
    :return List of translated images, each represented an a ndarray of dtype=np.uint8 [0-255]
    '''
    X_imgs = np.array(X_imgs, dtype=np.float32)
    offsets = np.zeros([len(X_imgs), 2], dtype=np.float32)
    n_translations = 4
    X_translated_data = []

    tf.reset_default_graph()
    with tf.Session() as session:
        for i in range(n_translations):
            X_translated =\
                np.zeros([len(X_imgs), IMAGE_SIZE, IMAGE_SIZE, 3], dtype=np.float32)
                
            base_offset, size, w_start, w_end, h_start, h_end = get_translate_params(i)
            offsets[:,:] = base_offset
            glimpses = tf.image.extract_glimpse(X_imgs, size, offsets)
            glimpses = session.run(glimpses)
            
            X_translated[:, h_start:h_end, w_start:w_end, :] = glimpses
            X_translated_data.extend(X_translated)
     
    X_translated_data = np.array(X_translated_data, dtype=np.uint8)
    return X_translated_data

def finer_rotate_images(X_imgs):
    '''
    Performs image rotation in the 2D plane
    :param X_imgs: List of images represented as a ndarray
    :return List of rotated images, each represented an a ndarray of dtype=np.uint8 [0-255]
    '''
    tf.reset_default_graph()
    
    X_rotate_data = []
    X = tf.placeholder(tf.float32, shape=[None, IMAGE_SIZE, IMAGE_SIZE, 3])
    radian = tf.placeholder(tf.float32, shape=[len(X_imgs)])
    tf_img = tf.contrib.image.rotate(X, radian)
    
    n_rotations = np.linspace(-15, 15, 6)
    with tf.Session() as session:
        for degree in n_rotations:
            radian_val = degree * np.pi / 180
            radian_arr = [radian_val] * len(X_imgs)
            rotated_imgs = session.run(tf_img, feed_dict={
                X: X_imgs, radian: radian_arr
            })
            X_rotate_data.extend(rotated_imgs)

    X_rotate_data = np.array(X_rotate_data, dtype=np.uint8)
    return X_rotate_data

def add_gaussian_noise(X_imgs):
    '''
    Adds random gaussian noise to the images to simulate poor lighting conditions
    :param X_imgs: List of images represented as a ndarray
    :return List of scaled images, each represented an a ndarray of dtype=np.uint8 [0-255]
    '''
    gaussian_noise_imgs = []
    row, col, _ = X_imgs[0].shape
    
    # Gaussian distribution params
    mean = 0.0; var = 0.1; std = var ** 0.5
    for X_img in X_imgs:
        X_img = X_img.astype(np.float32)
        gaussian = np.random.random((row, col, 1)).astype(np.float32)
        gaussian = np.concatenate((gaussian, gaussian, gaussian), axis = 2)
        gaussian_img = cv2.addWeighted(X_img, 0.75, 0.25 * gaussian, 0.25, 0)
        gaussian_noise_imgs.append(gaussian_img)

    gaussian_noise_imgs = np.array(gaussian_noise_imgs, dtype = np.uint8)

    return gaussian_noise_imgs

def motion_blur(X_imgs):
    '''
    Simulates motion blurring of the images
    :param X_imgs: List of images represented as a ndarray
    :return List of scaled images, each represented an a ndarray of dtype=np.uint8 [0-255]
    '''
    X_out = []
    
    # generating the kernel
    size = 4
    kernel = np.zeros([size, size])
    kernel[int((size-1)/2), :] = np.ones(size)
    kernel = kernel / size
    
    for idx, img in enumerate(X_imgs):
        X_out.append(cv2.filter2D(img, -1, kernel))
    return np.array(X_out, dtype=np.uint8)

def augment_data(X_data, y_data):
    '''
    Generates additional images by leveraging various data augmentation techniques with the 
    aim of achieving uniform label distribution and a more diverse data set.
    :param X_imgs: List of images represented as a ndarray
    :return List of scaled images, each represented an a ndarray of dtype=np.uint8 [0-255]
    '''
    
    print('Starting the image augmentation process...')
    X_data_aug = []
    y_data_aug = []
    
    # Compute the histogram of a set of data
    # NOTE: the kwarg 'density=True' returns value of the PDF over each bin
    hist, _ = np.histogram(y_data, n_classes, density=True)
    
    # Determine labels that fall below the threshold. It's these labels for which
    # we perform data augmentation
    labels_to_augment = np.where(hist < 0.02)[0]
    
    for label in labels_to_augment:
        images = X_data[y_data == label]
        gen_images = []
        
        for img in tqdm(images, desc='Label {}'.format(label)):
            aug_type = np.random.randint(0, 3)
            if aug_type == 0:
                gen_images.extend(centrally_scale_images([img]))
            elif aug_type == 1:
                gen_images.extend(translate_images([img]))
            elif aug_type == 2:
                gen_images.extend(finer_rotate_images([img]))
            elif aug_type == 3:
                gen_images.extend(add_gaussian_noise([img]))
        
        print('Generated {} new images for label {}.'.format(len(gen_images), label))
        X_data_aug.extend(gen_images)
        y_data_aug.extend([label] * len(gen_images))
    
    # Increasing the data diversity by adding motion blurring to each image from the
    # original data set
    print('Adding motion blur to images from the original data set...')
    X_data_mb = motion_blur(X_data)
    y_data_mb = y_data
    print('Number of generated mb images/labels: {}/{}'.format(len(X_data_mb), len(y_data_mb)))
    X_data_aug.extend(X_data_mb)
    y_data_aug.extend(y_data_mb)

    print('Image augmentation process complete!')
    return np.array(X_data_aug, dtype=np.uint8), np.array(y_data_aug, dtype=np.uint8)
````

````python
#Pre-process the data set to achieve feature value range of [-1, 1]
from skimage import exposure

def pre_process_image(img):
    '''
    Pre-processes the passed image. The steps include:
    RGB -> Grayscale -> Feature Scaling + Mean Normalisation -> Contrast Limited Adaptive Histogram Equalisation
    :param img: Image, represented as a ndarray of dtype=uint8
    :return Pre-processed image
    '''
    img_p = cv2.cvtColor(img, (cv2.COLOR_BGR2YUV))[:,:,0]
    img_p = (img_p / 255.0).astype(np.float32) # [0-1] interval
    img_p = (exposure.equalize_adapthist(img_p) - 0.5) # [-0.5, +0.5] interval

    return img_p

def pre_process_images(X):
    print('Started the pre-processing process...')
    print('Input shape: {}'.format(X.shape))
    X_out = np.empty([X.shape[0], X.shape[1], X.shape[2]]).astype(np.float32) 
    print('Output shape: {}'.format(X_out.shape))
    
    for idx, img in tqdm(enumerate(X)):
        X_out[idx] = pre_process_image(img)
    print('Completed the datset pre-processing.')
    return X_out

# Perform normalisation over the training/validation/test set 
X_train = pre_process_images(X_train)
X_valid = pre_process_images(X_valid)
X_test  = pre_process_images(X_test)

# Placeholder for alternative image pre-processing techniques
# def mean_normalisation(data):
#     image_values = np.arange(0, 256)
#     mu, sigma = np.mean(image_values), np.std(image_values)
#     data = (data - mu) / sigma
#     return data    

# # Convert the color space for the Training set from RGB -> Gray scale
# def rgb2gray(X):
#     X = 0.2989 * X[...,0] + 0.5870 * X[...,1] + 0.1140 * X[...,2]
#     return X
````

````python
# Load the pickled data
data_file = './preprocessed_dataset.p'
with open(data_file, mode='rb') as f:
    data = pickle.load(f)
    
X_train, y_train = data['features_train'], data['labels_train']
X_valid, y_valid = data['features_valid'], data['labels_valid']
X_test, y_test = data['features_test'], data['labels_test']

# Sanity checks
assert(len(X_train) == len(y_train))
assert(len(X_valid) == len(y_valid))
assert(len(X_test) == len(y_test))
````

````python
# Defining the model's Hyper parameters
from tensorflow.contrib.layers import flatten

tf.reset_default_graph()

EPOCHS = 20
BATCH_SIZE = 128
N_LABELS = n_classes
LEARNING_RATE = 0.001
LAMBDA = 0.001 # L2 regularisation loss term
# Values for weight init
mu = 0
sigma = 0.1

X = tf.placeholder(tf.float32, [None, IMAGE_SIZE, IMAGE_SIZE])
y = tf.placeholder(tf.int32, [None])
one_hot_y = tf.one_hot(y, N_LABELS)
learning_rate = tf.placeholder(tf.float32)
apply_dropout = tf.placeholder(tf.bool)
keep_prob = tf.placeholder(tf.float32, shape=[4])

# Helper functions
def kaiming_he_init(weight_shape, layer_type='fc'):
    '''
    Perform weight initialisation using the Kaiming He technique as explained in
    http://arxiv-web3.library.cornell.edu/abs/1502.01852
    :param weight_shape: Shape of the resulting weight matrix
    :param layer_type: Layer type ('conv'/'fc') This is required to correctly perform the weight init
    :return Tensor variable with shape 'weight_shape' and with init value as obtained through the
    Kaiming He technique
    '''
    fan_in, init_value = None, None
    
    if layer_type == 'conv':
        fan_in = weight_shape[0] * weight_shape[1] * weight_shape[2]
        init_value = tf.truncated_normal(weight_shape) * tf.sqrt(2 / fan_in)
    elif layer_type == 'fc':
        fan_in = weight_shape[0]
        init_value = tf.truncated_normal(weight_shape) * tf.sqrt(2 / fan_in)
    
    return tf.Variable(init_value)

layer_depth = {
    'c1': 16,
    'c2': 32,
    'c3': 32,
    'fc1': 320,
    'out': N_LABELS
}

weights = {
    'c1': kaiming_he_init((5, 5, 1, layer_depth['c1']), 'conv'),
    'c2': kaiming_he_init((5, 5, layer_depth['c1'], layer_depth['c2']), 'conv'),
    'c3': kaiming_he_init((3, 3, layer_depth['c2'], layer_depth['c3']), 'conv'),
    'fc1': kaiming_he_init((1584, layer_depth['fc1'])),
    'out': kaiming_he_init((layer_depth['fc1'], layer_depth['out']))
}

biases = {
    'c1': tf.Variable(tf.zeros(layer_depth['c1'])),
    'c2': tf.Variable(tf.zeros(layer_depth['c2'])),
    'c3': tf.Variable(tf.zeros(layer_depth['c3'])),
    'fc1': tf.Variable(tf.zeros(layer_depth['fc1'])),
    'out': tf.Variable(tf.zeros(layer_depth['out']))
}

def conv2d(input_vol, W, b, stride=1, padding='VALID'):
    conv_layer = tf.nn.conv2d(input_vol, W, strides=[1, stride, stride, 1], padding=padding)
    conv_layer = tf.nn.bias_add(conv_layer, b)
    return tf.nn.relu(conv_layer)

def maxpool2d(input_vol, k=2, stride=2):
    return tf.nn.max_pool(input_vol, ksize=[1, k, k, 1],
                         strides=[1, k, k, 1],
                         padding='VALID')

### Implement the model architecture
def LeNet(X, weights, biases, apply_dropout):
    '''
    Builds a CNN inspired by the model presented by Pierre Sermanet and Yann LeCun
    in their paper http://yann.lecun.com/exdb/publis/pdf/sermanet-ijcnn-11.pdf 
    '''
    if apply_dropout is True:
        print("Training phase -> perform Dropout")
    else:
        print("Evalutation phase -> not performing Dropout") 
    
    # Layer 1: Convolutional/Pooling/Dropout. 
    X = tf.expand_dims(X, -1)    
    conv1 = conv2d(X, weights['c1'], biases['c1']) # Input = 32x32x1. Output = 28x28x16.
    conv1 = maxpool2d(conv1)                       # Input = 28x28x16. Output = 14x14x16.
    conv1 = tf.cond(apply_dropout, lambda: tf.nn.dropout(conv1, keep_prob=keep_prob[0]), lambda: conv1)

    # Layer 2: Convolutional/Pooling/Dropout. 
    conv2 = conv2d(conv1, weights['c2'], biases['c2']) # Input = 14x14x16. Output = 10x10x132.
    conv2 = maxpool2d(conv2)                           # Input = 10x10x32. Output = 5x5x32.
    conv2 = tf.cond(apply_dropout, lambda: tf.nn.dropout(conv2, keep_prob=keep_prob[1]), lambda: conv2)
    
    # Layer 3: Convolutional/Dropout. 
    conv3 = conv2d(conv2, weights['c3'], biases['c3'], padding='SAME') # Input = 5x5x32. Output = 5x5x32.
    conv3 = tf.cond(apply_dropout, lambda: tf.nn.dropout(conv3, keep_prob=keep_prob[2]), lambda: conv3)

    # Intermediate layer: Multi-scale features/Pooling
    conv1_1 = maxpool2d(conv1)     # Input = 14x14x16. Output = 7x7x16
    shape = conv1_1.get_shape().as_list()
    conv1_1 = tf.reshape(conv1_1, [-1, shape[1] * shape[2] * shape[3]]) # Input = 7x7x16. Output = Nonex784. 

    # Flatten conv3
    shape = conv3.get_shape().as_list()
    conv3 = tf.reshape(conv3, [-1, shape[1] * shape[2] * shape[3]]) # Input = 5x5x32. Output = Nonex800
    
    # Layer 4: Fully Connected/Dropout.
    fc1 = tf.concat([conv1_1, conv3], 1)  # Input = 784+800 = 1584. Output = 320.
    fc1 = tf.matmul(fc1, weights['fc1']) + biases['fc1']
    fc1 = tf.nn.relu(fc1)
    fc1 = tf.cond(apply_dropout, lambda: tf.nn.dropout(fc1, keep_prob=keep_prob[3]), lambda: fc1)
    
    # Layer 5: Fully Connected
    logits = tf.matmul(fc1, weights['out']) + biases['out']
    return logits

# Training pipeline
logits = LeNet(X, weights, biases, apply_dropout)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=one_hot_y)
loss_data = tf.reduce_mean(cross_entropy)
loss_regularisation = tf.reduce_sum([tf.nn.l2_loss(w) for w in weights.values()])

loss = tf.reduce_mean(loss_data + LAMBDA * loss_regularisation)
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)

### Model evaluation pipeline
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    session = tf.get_default_session()
    
    for offset in range(0, num_examples, BATCH_SIZE):
        X_batch, y_batch = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = session.run(accuracy_operation, feed_dict={
            X: X_batch, 
            y: y_batch, 
            apply_dropout: False, 
            keep_prob: [1.0, 1.0, 1.0, 1.0]})
        
        total_accuracy += (accuracy * len(X_batch))
    return total_accuracy / num_examples
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
global best_validation_accuracy
best_validation_accuracy = 0.0
````

````python
# Start a tf.Session() and run the training + evaluation pipeline to learn the network weights/biases
with tf.Session() as session:
    session.run(tf.global_variables_initializer())
#     saver.restore(session, './ckpts/lenet11_pmbt')

    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        
        # Perform learning rate decay. Reduce the LEARNING_RATE by factor of 10
        # everytime the validation accuracy saturates
#         if i > 10:
#             LEARNING_RATE = 0.0001
            
        for offset in tnrange(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            X_batch, y_batch = X_train[offset:end], y_train[offset:end]
            session.run(optimizer, feed_dict={X: X_batch, y: y_batch, apply_dropout: True,
                                                   learning_rate: LEARNING_RATE, keep_prob:[0.8,0.8,0.8,0.8]})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        training_accuracy = evaluate(X_train, y_train)
        print("EPOCH {} ...".format(i+1))
        print("Training Accuracy = {:.5f}".format(training_accuracy))
        print("Validation Accuracy = {:.5f}".format(validation_accuracy))
        print()
        if (validation_accuracy > best_validation_accuracy):
            best_validation_accuracy = validation_accuracy
            saver.save(session, './ckpts/lenet11_pmbt')
            print("Model saved")

print('Training complete!')
````

````python
# Evaluate the final Training/Validation/Test accuracy of the model.

ckpt_path = './ckpts/lenet11_pmbt'
with tf.Session() as session:
    saver.restore(session, ckpt_path)
    training_accuracy = evaluate(X_train, y_train)
    validation_accuracy = evaluate(X_valid, y_valid)
    test_accuracy = evaluate(X_test, y_test)
    print("Training Acccuracy: {:.4f}".format(training_accuracy))
    print("Validation Acccuracy: {:.4f}".format(validation_accuracy))
    print("Test Acccuracy: {:.4f}".format(test_accuracy))
````
